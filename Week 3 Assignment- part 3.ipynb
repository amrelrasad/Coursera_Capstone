# Part 1: Read and process data from web

# 1. Read table from `wikipedia` page using `pandas` `read_html` function

```python
page = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'    
wikitables = pd.read_html(page, index_col=0, attrs={"class":"wikitable"})     
#wikitables[0].to_csv('assignment1.csv') 
```

```python
#read data frame from table
df_can_n=wikitables[0]
```

# 2. Process Dataframe

```python
#show head of dataframe
df_can_n.head()
```

```python
#rest index to swith postcode ino columns
df_can_n.reset_index( inplace=True)
#rename columns
df_can_n.columns=df_can_n.iloc[0,:]
#drop first row
df_can_n.drop(index=0,inplace=True)
# reset index to ensure continunity after dropped rows
df_can_n=df_can_n.reset_index(drop=True)
# drop not assigned Borough cells
df_can_n=df_can_n[df_can_n['Borough']!='Not assigned'].reset_index(drop=True)
# group neighbourhoods by postcode
df_can_n = df_can_n.groupby(['Postcode','Borough'])['Neighbourhood'].apply(', '.join).reset_index()
# dictionary for last row
last_row={'Postcode':df_can_n.shape[0],'Borough':'','Neighbourhood':''}
# append last row
df_can_n=df_can_n.append(last_row,ignore_index=True)
```

# Part 3: Clustering and generate map


```python
#imports
from sklearn.cluster import KMeans
import folium
```

```python
# new data frame for Toronto
df_toronto = df_can_n.copy()
df_toronto = df_toronto[df_can_n.Borough.str.contains("Toronto")]
df_toronto.reset_index(inplace=True,drop=True)
df_toronto.head()
```

```python
toronto_map = folium.Map(location=[43.65, -79.4], zoom_start=12)

x = df_toronto['Latitude']
y = df_toronto['Longitude']
packed_data = np.stack((x, y), axis=1)

kmeans = KMeans(n_clusters=4, random_state=0).fit(packed_data)

clusters = kmeans.labels_
colors = ['blue', 'red', 'green', 'black']
df_toronto['Cluster'] = clusters

for latitude, longitude, borough, cluster in zip(df_toronto['Latitude'], df_toronto['Longitude'], df_toronto['Borough'], df_toronto['Cluster']):
    label = folium.Popup(borough, parse_html=True)
    folium.CircleMarker(
        [latitude, longitude],
        radius=8,
        popup=label,
        color='yellow',
        fill=True,
        fill_color=colors[cluster],
        fill_opacity=0.9).add_to(toronto_map)  

toronto_map
```
